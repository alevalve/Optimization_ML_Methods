## Implementing the Gradient Descent Algorithm 

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
## Generate and Plot Cost Function 

## Generate random variables
theta = np.arange(24) 
#quadratic cost function 
J = (theta - 12)**2 
print(theta,J)
# Finding the value of x that minimizes J
plt.style.use('dark_background')
plt.plot(theta,J)
# adding labels to the curve
plt.ylabel("J (θ)")
plt.xlabel("θ")
plt.title("Cost Function")
# printing the plot
plt.show()
```

```python
## Optimize Cost Function to find the minima

def optimization (actual, learn_rate):
    error=[]
    for i in range (50): ## Quantity of iterations over the cost function
        gradient = 2*(actual - 10) ## Derivative respect to beta to obtain the gradient
        actual  = actual - learn_rate*gradient ## Compute the new location of θ (θ - lr * ∇J(θ))
        J = (actual - 10 )**2 ##  Equation to compute the error
        error.append(J) 
        print(actual) ## The convergent value for the interception is almost 10.
        #print(error)
        
    plt.plot(error) ## The squared error was reduced from 65 to near 0 in around 11-15 iterations.
    plt.title("Error value after 50 iterations")
    plt.xlabel("Iterations")
    plt.ylabel("Error")
    plt.show()
   
        
optimization(0,0.1)
        
# actual = 0
# learn_rate = 0.1
```
