Quasi Newthon Method Example

```python
## Import libraries
import numpy as np
import pandas as pd
from scipy.optimize import minimize
import matplotlib.pyplot as plt

```

```python
##  Define a function (X^2 - Y^2 + 5X)
x0 = np.array([1450, 2350, 3453, 1500, 1234]) ## Initial values

def f(x):
    return sum(x**3 + x**2 - 5*x) ## original function 

def dx(x):
    return (3*x**2 + 2*x + 5) ## derivative of the function 

```

```python
## Plot functions

x = np.linspace(-8, 5, 100) ## Original function (Blue)
plt.plot(x, f(x));

xo = np.linspace(-8, 5, 100) ## Derivative of the function (Orange)
plt.plot(x, dx(x));
```

```python
## Print BFGS optimization
res = minimize(f, x0, method='BFGS', jac=dx,
             options={'disp': True}) 
res.x 

## For this group of X0 we found a current fuction value of -13.61, the "desired" value is 0, therefore.
## It can be say that BFGS with these params and initial values hasnt found correctly the minimum after 23 iterations.
```
